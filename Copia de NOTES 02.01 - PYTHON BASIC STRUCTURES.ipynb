## `for` loops

unlike other languages, `for` loops are defined over **iterables**. A `list` is an iterable, and there are many other iterables.

Python syntax is **indented**.

Observe python determines the semantics of `*` according to the context

Sobre un iterable

```python
b = [1,10, 20.5, "hola", 10.]
```

Recorre los valores de b

```python
for i in b:
    print (i, "-->", type(i), "x2 -->", i*2)
```

another example of python as interpreted language

```python
b = [1,10, 20.5, "hola", 10.]
for i in b:
    print (i, "-->", type(i), "x2 -->", i**2)
```

**iterators**

sometimes we do not need to hold all the contents of a list in memory, but they can be generated as they are being asked for.

range cada vez que se le pide un elemnto lo genera

```python
k = range(-3,10,2)
k
```

size in memory

```python
k.__sizeof__()
```

```python
for i in k:
    print (i, end=", ")
```

Range genera pero no almacena a menos que se ponga

big_range = list(range(0,100000,2))

```python
big_range = range(0,100000,2)
big_range.__sizeof__()
```

```python
s = 0
for i in big_range:
    s += i
s
```

if converted to a list, then they are physically in memory

```python
a = list(big_range)
a.__sizeof__()
```

zip iterators: recorre los dos iteradores simultaneamente hasta que alguno se agote

```python
a = ["nada", "adios", 10.]
b = [1,10, 20.5, "hola", 10.]

for i,j in zip(a,b):
    print (i,j)
```

```python
for idx, elem in enumerate(a):
  print("elem at index", idx, "is", elem)
```

## Other control structures

string formatting / if / then / break / continue / while, etc.

```python
b = [1,10, 20.5, "hola", 10., 12]
for i in b:
    if i=='hola':
        break
    print (i, type(i))
```

```python
b = [1,10, 20.5, "hola", 10., 12]
for i in b:
    if i=='hola':
        break
    elif type(i)==int:
        print ("INT", end=" ")
    elif i>10:
        print (">10", end=" ")
    print (i, type(i))
```

```python
for i in b:
    if i=='hola':
        continue
    elif type(i)==int:
        print ("INT", end=" ")
    elif i>10:
        print (">10", end=" ")
    else:
        print ("UNK", end=" ")
    print (i, type(i))        
    
```

```python
for pos, i in enumerate(b):
    if i=='hola':
        continue
    print ("%02d :: %5.1f :: %20s"%(pos,i,type(i)))
```

```python
i=1
while i<len(b):
    print (i, b[i], type(b[i]))
    i += 1
```

## Dictionaries

Para definir un diccionario se usan los {}

```python
d[20]
```

```python
d.values()
```

```python
d.keys()
```

```python
for i in d.keys():
    print("la clave", i, "tiene el valor", d[i])
```

```python
for k,v in d.items():
    print("la clave", k, "tiene el valor", v)    
```

Las claves solo pueden ser objetos inmutables, una cadena o un número son inmutables

```python
d[48] = "otro numero"
d["nada"] = 0
```

```python
for k,v in d.items():
    print("la clave", k, "tiene el valor", v)    
```
